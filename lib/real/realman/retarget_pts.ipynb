```python
import numpy as np
from sklearn.linear_model import LinearRegression

XZ = np.array([
    [77.31, 179.90],
    [68.67, 123.85],
    [62.09, 62.99],
    [61.49, 7.51],
    [285.27, 157.27],
    [284.44, 113.24],
    [282.69, 63.13],
    [283.34, 16.74],
    [430.28, 130.65],
    [433.87, 94.36],
    [436.13, 54.27],
    [442.69, 18.94],
])

yz = np.array([
    [0.2, 0.35],
    [0.2, 0.3],
    [0.2, 0.25],
    [0.2, 0.2],
    [0.0, 0.35],
    [0.0, 0.3],
    [0.0, 0.25],
    [0.0, 0.2],
    [-0.2, 0.35],
    [-0.2, 0.3],
    [-0.2, 0.25],
    [-0.2, 0.2],
])

model_y = LinearRegression().fit(XZ, yz[:, 0])  # y = f(X, Z)
model_z = LinearRegression().fit(XZ, yz[:, 1])  # z = f(X, Z)

def predict_yz(X, Z):
    input_array = np.array([[X, Z]])
    y_pred = model_y.predict(input_array)[0]
    z_pred = model_z.predict(input_array)[0]
    return y_pred, z_pred
```

```python
X_test = 341.84
Z_test = 121.07

y, z = predict_yz(X_test, Z_test)
print(f"Input: X={X_test}, Z={Z_test}")
print(f"Prediction: y={y:.3f}, z={z:.3f}")
```
